### Detecting MULTICOLINEARITY using VIF

```python
# VIF handles multicolinerity from the dataset
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(dataset):
     vif = pd.DataFrame()
     vif["variables"]=dataset.columns
     vif["VIF"]=[variance_inflation_factor(dataset.values ,col_index)  for col_index in range(len(dataset.columns))]
     return vif

'''
variance_inflation_factor(ndarray, column_index_int)
'''

calc_vif(dataset[quan])
```

CODE EXPLANATION:-

I calculated the Variance Inflation Factor (VIF) to check for multicollinearity among the quantitative variables in the dataset.
First, I created a DataFrame containing only the numerical (quantitative) columns. Then, I computed the VIF for each column using list comprehension.

I iterated through each column index and passed two required parameters to the variance_inflation_factor() function:

exog – the independent variables in array format (dataset.values)

exog_idx – the index of the column for which VIF is being calculated

Since the function requires these two parameters according to the documentation, I passed the dataset values and the respective column index inside the loop.

ssc_p	hsc_p	degree_p	etest_p	mba_p	


These variables had VIF values greater than 5, indicating high multicollinearity.


The Salary column had a VIF of 3.93, which is less than 5, indicating moderate correlation and acceptable levels of multicollinearity.


Therefore, I need to handle the highly correlated features

```python
''' these numbers are simply NumPy’s scientific notation for floating-point numbers.  '''
dataset[quan].values
```

### HANDLING MULTICOLINEARITY

Therefore, I need to handle the highly correlated features by either removing some variables, applying dimensionality reduction techniques (such as PCA), or using regularization methods to reduce multicollinearity.

ways:-
1. Common remedies include removing one of the highly correlated variables.
2. using regularization techniques like Ridge or Lasso regression to stabilize coefficients
3. Feature Engineering (Combine Features)
4. Principal Component Analysis (PCA)
5. Use Tree-Based Models

```python
# 1. Common remedies include removing one of the highly correlated variables
calc_vif(dataset[['etest_p','salary']])
```

```python
# using regularization techniques like Ridge or Lasso regression to stabilize coefficients
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
x=dataset[['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p', 'gender_M',
       'ssc_b_Others', 'hsc_b_Others', 'hsc_s_Commerce', 'hsc_s_Science',
       'degree_t_Others', 'degree_t_Sci&Tech', 'workex_Yes',
       'specialisation_Mkt&HR', 'status_Placed']]
y=dataset[['salary']]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20, random_state=0)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
from sklearn.linear_model import RidgeCV
ridge = RidgeCV(alphas=(0.1, 1.0, 10.0),cv=5)
ridge.fit(x_train,y_train)
y_pred = ridge.predict(x_test)
from sklearn.metrics import r2_score
r2_score = r2_score(y_test,y_pred)
print(r2_score)     #  88% accuracy 

```

## Residual Plot 

```python

import numpy as np
# Convert everything to 1D numpy arrays
y_test_array = np.array(y_test).reshape(-1)
y_pred_array = np.array(y_pred).reshape(-1)
# Now subtract
residuals = y_test_array - y_pred_array
print("y_pred shape:", y_pred_array.shape)
print("residuals shape:", residuals.shape)

import matplotlib.pyplot as plt
plt.figure(figsize=(6,5))
plt.scatter(y_pred_array, residuals, alpha=0.6)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel("Fitted Values (Predicted)")
plt.ylabel("Residuals")
plt.title("Residuals vs Fitted Plot")
plt.show()
```

'''
I plotted the residuals for the Ridge Regression model. The residual plot shows that errors slightly increase as predicted values increase, indicating mild heteroscedasticity. While Ridge regression stabilizes the model, this pattern suggests the variance is not fully constant. To correct this, we can apply transformations to the target variable, use Weighted Least Squares, or adopt robust regression techniques.
'''

### Homoscedasticity (Equal Variance):

Purpose: Ensures that the variance of the errors (residuals) is constant across all levels of the independent variable. 

Use: It is a foundational assumption of linear regression. 

### Heteroscedasticity (Unequal Variance):

Purpose: Identifies that the spread of residuals changes (e.g., increases or decreases) with the level of the independent variable, signaling a potential model issue.

Use: It highlights when OLS estimates are no longer the best linear unbiased estimators (BLUE).

Handled by : Transformation, WLS, robust regression, tree based models

## T test 

### unparied test - independant sample 
different group (male,female - categorical) and same condition (salary -numerical)

```python
from scipy.stats import ttest_ind
male=dataset[dataset['gender']=='M']['salary']
female=dataset[dataset['gender']=='F']['salary']
ttest_ind(male,female)
```

'''pvalue=0.035995606467451834  (low)      similar groups'''

high similarity

### paried test - dependant sample 
same group (male - categorical) and different condition (ssc_p, hsc_p -numerical)

```python
from scipy.stats import ttest_ind
male_ssc=dataset[dataset['gender']=='M']['ssc_p']
male_hsc=dataset[dataset['gender']=='M']['hsc_p']
ttest_ind(male_ssc,male_hsc)
```

'''pvalue=0.644060993886658 (high)      different groups'''

low similarity

```python
# unpaired test - different group (ssc_b others,central - categorical) and same condition (salary -numerical)
from scipy.stats import ttest_ind
others=dataset[dataset['ssc_b']=='Others']['salary']
central=dataset[dataset['ssc_b']=='Central']['salary']
print("pvalue=0.6139510180995318    --> different groups ")
ttest_ind(others,central)
```

```python
from scipy.stats import ttest_ind
others=dataset[dataset['hsc_s']=='Commerce']['mba_p']
central=dataset[dataset['hsc_s']=='Science']['mba_p']
print("pvalue=0.4643299525385431   --> different groups ")
ttest_ind(others,central)
```

```python
# paired test - same group (commerce - categorical) and different condition (mba_p,etest_p -numerical)
from scipy.stats import ttest_ind
commerce_mba = dataset[dataset['hsc_s']=='Commerce']['mba_p']
commerce_etest =  dataset[dataset['hsc_s']=='Commerce']['etest_p']
print("pvalue=1.8442994548708624e-11 (0.000000000018442994548708624)    --> same groups")
ttest_ind(commerce_mba,commerce_etest)
```
